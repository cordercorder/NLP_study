```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
def get_page_source_code(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'lxml')
    return soup
```

```python
x = get_page_source_code('http://maggch.cn:8000/')
```

```python
href = []
for item in x.find_all('a'):
    tmp = item.attrs.get('href', None)
    if tmp:
        href.append(tmp)
```
