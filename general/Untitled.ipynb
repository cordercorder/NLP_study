```python
x = np.linspace(-10, 10, 201)
y = 1/(1 + np.exp(-x))
x2 = np.linspace(-2, 2, 101)
```

```python
plt.title('sigmod(x) and x')
plt.axvline(x = 0, ls = '--', c = 'r')
plt.plot(x2, 0.25*x2+0.5, label='y=x')
plt.plot(x, y, label='sigmod(x)')
plt.legend()
```
